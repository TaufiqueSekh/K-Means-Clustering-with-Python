## K Means Cluster Creation


```python
from sklearn.cluster import KMeans
```

```python
kmeans=KMeans(n_clusters=2)
```

```python
kmeans.fit(df.drop('Private',axis=1))
```

```python
kmeans.cluster_centers_
```

# Evalution

There is no perfect way to evaluate clustering if we don't have the labels, however since this is just an exercise, we do have the labels, so we take advantage of this to evaluate our clusters, but, we usually won't have this luxury in the real world.


```python
df['Cluster'] = df['Private'].map({'Yes': 1, 'No': 0})
```

```python
df.head()
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print("Classification Report : \n",classification_report(df['Cluster'],kmeans.labels_))
print("Confusion Matrix : \n",confusion_matrix(df['Cluster'],kmeans.labels_))
```

From the above Elbow graph, let's choose n_clusters=6 and check classification_report

```python
kmean=KMeans(n_clusters=6)
kmean.fit(df.drop(['Private','Cluster'],axis=1))
print("Classification Report : \n",classification_report(df['Cluster'],kmean.labels_))
print("Confusion Matrix : \n",confusion_matrix(df['Cluster'],kmean.labels_))
```
